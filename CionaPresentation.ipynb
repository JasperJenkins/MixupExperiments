## Semantic Segmentation
- Pixelwise classification

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

mask_cmap = ListedColormap([(244/255, 66/255, 83/255, a) for a in np.linspace(0, 1, 255)])
img_a, img_b = plt.imread('./img_pair0.jpg')/255, plt.imread('./img_pair1.jpg')/255
mask_a, mask_b = plt.imread('./mask_pair0.jpg')[...,0]/255, plt.imread('./mask_pair1.jpg')[...,0]/255

f, ax = plt.subplots(1, 4)
f.set_figwidth(20)
f.set_figheight(20)
ax[0].imshow(img_a)
ax[2].imshow(img_b)
ax[1].imshow(img_a)
ax[1].imshow(mask_a, cmap=mask_cmap, alpha=.75)
ax[3].imshow(img_b)
ax[3].imshow(mask_b, cmap=mask_cmap, alpha=.75)
```

## Mixup Data Augmentation

### Emperical Risk Minimization
- Minimizing loss using only observed data

### Vicinal Risk Minimization
- Minimizing loss using data in the vicinity of observed data
- Flipping, scaling, rotating, color transformations, affine transformations, etc. on images

### [mixup: Beyond Empirical Risk Minimization](https://openreview.net/forum?id=r1Ddp1-Rb)
- Interoplation of samples

```python
import numpy as np

def interpolate(a, b, f):
  return (a * (1.0 - f)) + (b * f)

lerps = 7

f, ax = plt.subplots(1,lerps)
f.set_figwidth(40)
f.set_figheight(40)
for i in range(lerps):
  mixup_lerp = i / (lerps - 1)
  ax[i].imshow(interpolate(img_a, img_b, mixup_lerp))
  ax[i].imshow(interpolate(mask_a, mask_b, mixup_lerp), alpha=.75, cmap=mask_cmap)
plt.show()
```
